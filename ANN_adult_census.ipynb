```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Ani-94/Adult_census/main/adult.csv')
```

```python
df
```

```python
df.columns
```

```python
categorical_feature = [features for features in df.columns if df[features].dtype == 'O']
```

```python
categorical_feature
```

```python
val = df[categorical_feature[0]].unique()[5]
```

```python
val
```

```python
for i in categorical_feature:
    val1 = df[i].value_counts().sort_values(ascending=False).index[0]
    df_n = df[i].replace(val,val1)
    df[i] = df_n
```

```python
df
```

```python
df1= df.loc[:,df.columns=='salary']
```

```python
val = df1['salary'].unique()[0]
```

```python
df1['salary'] = (df1['salary'] != val).astype(int)
```

```python
df['salary'] = df1
```

```python
df
```

```python
categorical_feature = [features for features in df.columns if df[features].dtype == 'O']
```

```python
categorical_feature
```

```python
new_df = pd.get_dummies(df,columns=categorical_feature)
```

```python
new_df
```

```python
y = new_df.loc[:,new_df.columns == 'salary']
x = new_df.loc[:,new_df.columns != 'salary']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
xtrain=sc.fit_transform(X_train)
xtest=sc.transform(X_test)
```

```python
xtrain
```

```python
xtrain.shape
```

```python
xtest
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU, ELU, ReLU
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential ()
```

```python
classifier.add(Dense (104,input_shape=(105,), activation= 'relu'))
```

```python
classifier.add(Dense(units=104, activation= 'relu'))
```

```python
classifier.add(Dense(units=104, activation= 'relu'))
```

```python
classifier.add(Dense(units=104, activation='relu'))
```

```python
classifier.add(Dense(1, activation='sigmoid'))
```

```python
classifier.summary()
```

```python
classifier.compile(optimizer='adam',loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
early_stopping= tf.keras.callbacks.EarlyStopping(
monitor='val_loss',
min_delta= 0.0001,
patience= 50,
verbose =1,
mode='auto',
baseline = None,
restore_best_weights= False)
```

```python
xtrain.shape
```

```python
model_history= classifier.fit(xtrain,y_train,validation_data=(xtest,y_test),batch_size=13,epochs=100,callbacks=early_stopping)
```

```python
model_history.history.keys()
```

```python
y_pred = classifier.predict(xtest)
```

```python
y_p=np.round(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_p)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_p,y_test)
```
